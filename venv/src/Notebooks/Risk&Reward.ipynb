```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from datetime import datetime
import requests
```

```python
dataset = pd.read_csv('/home/prp1277/Github-Repos/finance/venv/src/data/msft_details.csv', parse_dates=True)
dataset
```

```python
column_names = list(dataset.columns)
column_names
```

```python
dataset['Datetime'] = pd.to_datetime(dataset['Date'].apply(str) + ' ' + dataset['Time'])
dataset
```

## A Basic Timeseries Chart

```python
msft_scatter = go.Figure()
msft_scatter.add_trace(
    go.Candlestick(
        x=dataset['Datetime'],
        open=dataset['Open'],
        high=dataset['High'],
        low=dataset['Low'],
        close=dataset['Close']
    )
)
msft_scatter.update_layout(
    title_text='Microsoft Returns',
    xaxis_rangeslider_visible=True
)
msft_scatter.show()
```

## Try Out an OHLC Series

```python
msft_OHLC = go.Figure(data=[go.Ohlc(
    x=dataset['Datetime'],
    open=dataset['Open'],
    high=dataset['High'],
    low=dataset['Low'],
    close=dataset['Close']
)])
msft_OHLC.show()
```
